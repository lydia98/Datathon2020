```
import pandas as pd
import numpy as np
import matplotlib as plt
```

```
genome_scores = pd.read_csv("genome-scores.csv")
genome_tags = pd.read_csv("genome-tags.csv")
```

```
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
genome_effective = genome_scores.sort_values(by=['movieId','relevance'], ascending=[True, False]).reset_index(drop=True)
genome_effective = genome_effective.drop(genome_scores[genome_scores['relevance'] < 0.5].index, inplace=False)
```

```
genome_effective['pair'] = tuple(zip(genome_effective['tagId'], genome_effective['relevance']))
genome_list = genome_effective.groupby('movieId')['pair'].apply(list).reset_index(name='tagSeries')
genome_list.to_csv('genome_list.csv')
```

```
movieIdList = genome_scores['movieId'].unique()
```

```
a = list(genome_tags['tagId'])
```

```
genome_score_tag_sort = genome_scores.sort_values(by=['movieId','tagId'])
genome_score_tag_sort
```

```
df = pd.DataFrame(0, index=movieIdList, columns = genome_tags['tagId'])
df
```

```
tag_num = len(genome_tags['tagId'])
idx = 0
for i in movieIdList:
  df.loc[i] = genome_score_tag_sort[idx * tag_num:(idx + 1) * tag_num]['relevance'].tolist()
  idx += 1
```

```
df
```

```
df['movie_id'] = df.index
df
```

```
test_dataset = pd.read_csv("test_dataset.csv")
train_dataset = pd.read_csv("train_dataset.csv")
train_dataset
```

```
train_joinned = pd.merge(df, train_dataset, on='movie_id', how='inner', suffixes=('_left','_right'))
test_joinned = pd.merge(df, test_dataset, on='movie_id', how='inner', suffixes=('_left','_right'))
```
