```
import pandas as pd
import numpy as np
import matplotlib as plt
```

```
genome_scores = pd.read_csv("genome-scores.csv")
genome_tags = pd.read_csv("genome-tags.csv")
```

```
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```
movieIdList = genome_scores['movieId'].unique()
```

```
a = list(genome_tags['tagId'])
```

```
genome_score_tag_sort = genome_scores.sort_values(by=['movieId','tagId'])
genome_score_tag_sort
```

```
df = pd.DataFrame(0, index=movieIdList, columns = genome_tags['tagId'])
df
```

```
tag_num = len(genome_tags['tagId'])
idx = 0
for i in movieIdList:
  df.loc[i] = genome_score_tag_sort[idx * tag_num:(idx + 1) * tag_num]['relevance'].tolist()
  idx += 1
```

```
df
```

```
df['movie_id'] = df.index
df
```

```
test_dataset = pd.read_csv("test_dataset.csv")
train_dataset = pd.read_csv("train_dataset.csv")
train_dataset
```

```
train_joinned = pd.merge(df, train_dataset, on='movie_id', how='inner', suffixes=('_left','_right'))
test_joinned = pd.merge(df, test_dataset, on='movie_id', how='inner', suffixes=('_left','_right'))
```

```
#test_joinned.to_csv("test_inner_joinned.csv")
#train_joinned.to_csv("train_inner_joinned.csv")
train_joinned.head(5)
train_joinned.dropna()
```

```
from sklearn.linear_model import LogisticRegression
X_train = train_joinned.iloc[:,0:1128]
y_train = train_joinned[['category']]
```

```
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```
X_test = test_joinned.iloc[:,0:1128]
y_test = test_joinned[['category']]
predictions = lr.predict(X_test)
predictions
```

```
from sklearn.metrics import classification_report
target_names = ['class 1', 'class 2', 'class 3','class 4']
print(classification_report(y_test, predictions, target_names=target_names))
```
